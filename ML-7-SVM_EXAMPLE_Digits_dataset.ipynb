```python
x=digits.data
y=digits.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=18)
```

```python
model1=SVC()
model1.fit(x_train,y_train)
model1.score(x_test,y_test)
```

```python
model2=SVC(C=2.5,gamma="scale")
model2.fit(x_train,y_train)
model2.score(x_test,y_test)
```

```python
model3=SVC(kernel='linear',C=3,gamma="auto")
model3.fit(x_train,y_train)
model3.score(x_test,y_test)
```

```python
model4=SVC(kernel="rbf",C=4,gamma="scale")
model4.fit(x_train,y_train)
model4.score(x_test,y_test)
```

```python
model5=SVC(kernel="poly",degree=2,C=0.5,gamma="scale")#C>0
model5.fit(x_train,y_train)
model5.score(x_test,y_test)
```

```python
model6=SVC(kernel="sigmoid",C=1)
model6.fit(x_train,y_train)
model6.score(x_test,y_test)
```

```python
from sklearn.decomposition import PCA#PRINCIPAL COMPONENT ANALYSIS
from mlxtend.plotting import plot_decision_regions
```

```python
pca=PCA(n_components=2)
x_train_pca=pca.fit_transform(x_train)
x_test_pca=pca.fit_transform(x_test)
```

```python
model1_pca=SVC()
model1_pca.fit(x_train_pca,y_train)
plot_decision_regions(x_train_pca,y_train,clf=model1_pca)
plot_decision_regions(x_test_pca,y_test,clf=model1_pca)
plt.figure()
```

```python
model1_pca.score(x_test_pca,y_test)
```

```python
model2_pca=SVC(C=2.5,gamma="scale")
model2_pca.fit(x_train_pca,y_train)
plot_decision_regions(x_train_pca,y_train,clf=model2_pca)
plot_decision_regions(x_test_pca,y_test,clf=model2_pca)
plt.figure()
```

```python
model2_pca.score(x_test_pca,y_test)
```

```python
model3_pca=SVC(kernel='linear',C=3,gamma="auto")
model3_pca.fit(x_train_pca,y_train)
plot_decision_regions(x_train_pca,y_train,clf=model3_pca)
plot_decision_regions(x_test_pca,y_test,clf=model3_pca)
plt.figure()
```

```python
model3_pca.score(x_test_pca,y_test)
```

```python
model4_pca=SVC(kernel="rbf",C=4,gamma="scale")
model4_pca.fit(x_train_pca,y_train)
plot_decision_regions(x_train_pca,y_train,clf=model4_pca)
plot_decision_regions(x_test_pca,y_test,clf=model4_pca)
plt.figure()
```

```python
model4_pca.score(x_test_pca,y_test)
```
